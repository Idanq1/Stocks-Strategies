```python
from pandas_datareader import data
import matplotlib.pyplot as plt
import pandas as pd
import yfinance
```

```python
start_time = "2015-01-02"
end_time = "2022-10-17"
symbol = "BA"
```

```python
df = data.DataReader(symbol, "yahoo", start_time, end_time)
df
```

```python
df = yfinance.download(symbol, start_time, end_time)
df
```

```python
symbols = ["BA", "TSLA", "BTC-USD"]
```

```python
df = data.DataReader(symbols, "yahoo", start_time, end_time)
df
```

```python
closes = df.Close.copy()
```

```python
closes
```

```python
closes["TSLA"].dropna().plot(figsize=(15, 8), fontsize=13)
plt.legend(fontsize=13)
plt.show()
```

```python
closes.BA.div(closes.iloc[0, 0]).mul(100)  # Normalize to 100
```

```python
norm = closes.div(closes.iloc[0]).mul(100)
norm
```

```python
norm.dropna().plot(figsize=(15, 8), fontsize=13, logy = True)
plt.legend()
plt.show()
```

```python
btc = closes['BTC-USD'].dropna().to_frame().copy()
btc
```

```python
btc["returns"] = btc.diff()
btc
```
