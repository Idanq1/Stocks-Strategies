```python
with open(r"Energy tickers.txt", 'r') as f:
    energy_tickers = f.read().splitlines()
energy_tickers
```

```python
df = yf.download(energy_tickers, period="3y", interval="1d")
```

```python
tsla = yf.Ticker("ENB").history(period="3y", interval="1d")["Close"]
```

```python
df = df["Close"]
```

```python
df.drop("MPC", axis=1).mean(axis=1).pct_change().cumsum().plot(grid=True, label="Energy sector")
df["MPC"].pct_change().cumsum().plot(grid=True, label="MPC")
tsla.pct_change().cumsum().plot(grid=True, label="ENB")
plt.legend()
plt.show()
```

```python
top_energy_stocks = df.pct_change().cumsum().iloc[-1].sort_values(ascending=False).head(10).index.to_list()
```

```python
df[top_energy_stocks].mean(axis=1).pct_change().cumsum().plot(grid=True, label="Energy sector")
df["MPC"].pct_change().cumsum().plot(grid=True, label="MPC")
plt.legend()
```
